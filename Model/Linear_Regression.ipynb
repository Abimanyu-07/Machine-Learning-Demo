```python
import pandas as pd
```

```python
df = pd.read_csv('homeprices.csv')
df
```

Visualize Data

```python
import matplotlib.pyplot as plt

plt.scatter(df['area'],df['price'])
plt.xlabel('area')
plt.ylabel('price')
plt.title('Home Prices')
plt.show()
```

Linear Regression

```python
from sklearn.model_selection import train_test_split
X = df[['area']]
y = df['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print('Predicted:',y_pred)
print('Actual:',y_test.values)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print('MSE:',mean_squared_error(y_test,y_pred))
```

```python
predicted_price = model.predict(pd.DataFrame({'area': [3300]}))
print(f"Predicted price for a 3300 sq ft house: {predicted_price[0]:,.2f}")
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['area'], df['price'], color='red', marker='+')
plt.plot(df['area'], model.predict(df[['area']]), color='blue')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price (INR)')
plt.title('Home Prices Prediction (Linear Regression)')
plt.grid(True)
plt.show()
```
