```python
import pandas as pd
```

```python
df = pd.read_csv('dataset/homeprices.csv')
df
```

```python
X = df[['area']]
Y = df['price']

print("Features (X):")
display(X)
print("\nTarget (Y):")
display(Y)
```

Linear Regrssion

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(X, Y)

print(f"Coefficient (slope, m): {reg.coef_[0]:.2f}")
print(f"Intercept (b): {reg.intercept_:.2f}")
```

```python
predicted_price = reg.predict(pd.DataFrame({'area': [3300]}))
print(f"Predicted price for a 3300 sq ft house: {predicted_price[0]:,.2f}")
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['area'], df['price'], color='red', marker='+')
plt.plot(df['area'], reg.predict(df[['area']]), color='blue')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price (INR)')
plt.title('Home Prices Prediction (Linear Regression)')
plt.grid(True)
plt.show()
```
